Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(1, 30, 0.27)
y = 1.123*np.square(X) + 2.456

plt.figure(figsize=(6, 4))
plt.plot(X, y, c='r')
plt.xlabel("X")
plt.ylabel("y")
plt.title("X vs y")
plt.grid()
plt.show()
```

```python
# Plot using subplots
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.plot(X, y, c='r')
ax1.set_xlabel('X')
ax1.set_ylabel('y')
ax1.set_title('X vs y')

ax2 = fig.add_subplot(2,2,2)
ax2.plot(X, y, c='r')
ax2.set_xlabel('X')
ax2.set_ylabel('y')
ax2.set_title('X vs y')

plt.tight_layout()
```

Barchart:
    Used for descrete data

```python
fruits = ["Apples", "Bananas", "Oranges", "Gauva"]
quantity = [100, 120, 90, 130]

fig, axes = plt.subplots(2,2)
for i in range(2):
    for j in range(2):
        axes[i, j].bar(fruits, quantity, color = {"Red", "Yellow", "Orange", "Green"})
        axes[i, j].set_title(f"Plot {i} vs {j}")
        axes[i, j].set_xticks(ticks = fruits)
        axes[i, j].set_xticklabels(labels = fruits, rotation = 90)

plt.tight_layout()
# fig.subplots_adjust(wspace=1, hspace=1)
```

Histogram:
    Normality check, outlier detection, understanding data shape

```python
# Generating random data
data = np.random.normal(loc= 50, scale=10, size= 1000)

# Creating histogram
plt.hist(data, bins=20, color='magenta', edgecolor='black')

#Labels and title
plt.xlabel('Value')
plt.ylabel('Freq')
plt.title('Histogram of random data')
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X, y = make_moons(
    n_samples=1000,
    noise=0.03,
    random_state=42
)

fig, axes = plt.subplots()
axes.scatter(X[:,0], X[:,1], c=y)
print(X.shape)
print(y.shape)

axes.text(x=0, y=0, s="Manav Vai", family='monospace', fontsize=10)
```

```python
from sklearn.datasets import load_iris
```

```python
X, y = load_iris(return_X_y=True, as_frame=True)

X.head
```

```python
fig, axes = plt.subplots(2, 2)


axes[0, 0].bar(x= ['Setosa', 'Versicolor', 'Verginica'], height= y.value_counts(), color=['Red', 'Green', 'Blue'])

axes[0, 1].scatter(X['sepal length (cm)'], X['petal length (cm)'], c=y)

axes[1, 0].hist(X['sepal length (cm)'], bins=20, color='magenta', edgecolor='black')

axes[1, 1].boxplot(X['sepal length (cm)'])

plt.tight_layout()
```
