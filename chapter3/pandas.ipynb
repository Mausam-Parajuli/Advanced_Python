```python
import pandas as pd
```

```python
df_A = pd.DataFrame({
    'Temp': [20, 30, 40, 50]},
    index = ['day 1', 'day 2', 'day 3', 'day 4']
)
df_B = pd.DataFrame({
    'Humidity': [120, 130, 140, 150]},
    index = ['day 1', 'day 2', 'day 5', 'day 7']
)
```

```python
joined_df = df_A.join(df_B, how='left')
print(joined_df)
```

```python
# Melt converts columns into rows

wide_df = pd.DataFrame({
    'Name': ['Ram', 'Shyam', 'Hari', 'Sita'],
    'Maths': [10, 20, 30, 50],
    'Science': [30, 35, 34, 46],
    'English': [12, 25, 32, 12],
})
print(wide_df)
```

```python
long_df = pd.melt(
    frame= wide_df,
    id_vars= ['Name'],
    var_name= 'Subject',
    value_name= 'Marks'
)
print(long_df)
```

```python
# Pivot le long lai wide banauxa

feri_wide_df = long_df.pivot(
    index= 'Name',
    columns='Subject',
    values= 'Marks'
)
print(feri_wide_df)
```

```python
stacked_df = long_df.stack()
print(stacked_df)
```

```python
unstacked_df = stacked_df.unstack()
print(unstacked_df)
```

```python
df_pivot_table = pd.pivot_table(
    data= long_df,
    index= 'Name',
    columns='Subject',
    values= 'Marks',
    aggfunc='median'
)
print(df_pivot_table)
```
