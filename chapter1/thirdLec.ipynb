```python
# Class and Object, Inheritance

# Base class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f'{self.name} is Speaking')

# Derived class
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)    #super() le immediate parent class lai point garxa
        self.breed = breed

    def speak(self):
        super().speak()
        print(f"{self.name} of {self.breed} breed is Barking")

d1 = Dog("Lucy", "Doberman")
d1.speak()
```

```python
# Multiple Inheritance
class A:
    def show(self):
        print("A")

class B:
    def show(self):
        print("B")

class C(A, B):
    pass

c1 = C()
c1.show()
```

```python
# Method Resolution Order

C.mro()
```

```python
class DataList:
    def __init__(self, list_1):
        self.alist = list_1

    def mean(self):
        s = sum(self.alist)
        n = len(self.alist)
        return s / n
    
l1 = DataList([1,2,3,4,5,6])
l1.mean()
```

```python
# Exception handling, debugging and logging

try:
    n1 = int(input("First Number : "))
    n2 = int(input("Second Number : "))
    result = n1/n2
    print("Result is ", result)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Please enter valid number")
finally:
    print("This always runs")
```
