```python
word = "Hello World"
count = {'vowels': 0, 'consonants':0}
vowels = ['a','e','i','o','u','A','E','I','O','U']
for i in word:
    if i == " ":
        continue
    elif i not in vowels:
        count['consonants'] += 1
    else:
        count['vowels'] += 1
print(count)
```

```python
#Write a function that computes the average of jati ota numbers
#Args for unknown no of arguments
def average(*args):
    sum = 0
    for i in args:
        sum = sum + i
    return sum/len(args)

average(1,2,3,4,5)
```

```python
class Animal:
    def __init__(self,name):                 #Constructer, object banaune
        self.name = name 

    def __str__(self):                        #Object lai call garda, k dekhaune kura, aimed at users
        return f"Animal name is {self.name}"
    
    def __repr__(self):                        #Aimed at programmers, called when we just write an object
        return f"Animal({self.name})"          #ideally to recreate the object

    def sayHello(self):
        print(f"{self.name} says Hello!")

a1 = Animal("Tommy")
print(a1.name)
a1
```

```python
class Point:
    def __init__(self, xx, yy):
        self.x = xx
        self.y = yy

    def __str__(self):
        return f"{self.x, self.y}"
    
    def __add__(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x,new_y)
    
p1 = Point(1, 2)
p2 = Point(2,1)
p3 = p1 + p2     #p1.__add__(p2)
print(p3)
```

```python
# Home Work

# WAP to create a class named Distance which has
# feet and inches as data member. Overload any two operators

class Distance:
    def __init__(self, feet, inches):
        self.feet = feet
        self.inches = inches
    
    def __str__(self):
        return f"{self.feet} ft. {self.inches} inches"
    
    def __add__(self, other):
        new_ft = self.feet + other.feet
        new_in = self.inches + other.inches
        if new_in > 12:
            new_ft += int(new_in/12)
            new_in = int(new_in%12)
        return Distance(new_ft, new_in)
    
D1 = Distance(5, 7)
D2 = Distance(7, 9)
print(D1)
print(D2)
D3 = D1 + D2
print(D3)
```
