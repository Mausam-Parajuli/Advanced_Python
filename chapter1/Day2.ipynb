```python
def sum(n):
    s = 0
    for i in range(1,n+1):
        s += i
    return s

sum(1000)
```

```python
# Write an function named average which computes average
def average(*args):
    s = 0
    for num in args:
        s += num
    return s/len(args)


average(1,2,3,4,5,6,8,9,10)
```

```python
def myFunction(**kwargs):
    s = 0
    for key in kwargs.keys():
        s = s + kwargs[key]
    return s

myFunction(a=10,b=20)
```

```python
def anotherFunc(*args, **kwargs):
    print(args)
    print(kwargs)

anotherFunc(1,2,3,a=10, b=20)
```

```python
class Animal:
    def __init__(self,name):
        self.name = name

    def __str__(self):
        return f'Animal name is {self.name}'   
    
    def __repr__(self):
        return f'Animal : {self.name}'
    
    def sayHello(self):
        print(f"{self.name} says Hello!")

a1 = Animal("Tommy")
a1
```

```python
class Point:
    def __init__(self,xx,yy):
        self.x = xx
        self.y = yy 

    def __str__ (self):
        return f'({self.x}, {self.y})'
    
    def __add__(self,other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)
    
p1 = Point(1,2)
p2 = Point(2,1)
p3 = p1+p2 # p1.__add__(p2)
print(p3) 
```
