```python
import requests
import pandas as pd
from pandas import json_normalize

cities = ['Dharan', 'Kathmandu', 'Itahari', 'Biratnagar', 'Janakpur']
final_list = []

for city in cities:
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={key}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        df = json_normalize(data)

        df_weather = df[[
            'name',
            'main.temp',
            'main.humidity',
            'wind.speed',
            'wind.deg',
            'weather'
        ]]
        df_weather.head()

        df_weather['discription'] = df_weather['weather'].apply(lambda x: x[0]['description'])
        df_weather = df_weather.drop(columns=['weather'])
        print(df_weather.head())

        final_list.append(df_weather)

    else:
        print(f"{city} → Status Code: {response.status_code}")

# combine all cities into one dataframe
final_df = pd.concat(final_list, ignore_index=True)

print(final_df)
```
