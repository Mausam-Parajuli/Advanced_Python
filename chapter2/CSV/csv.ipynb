```python
import csv

data = [
    ["Name", "Address", "Age"],
    ["Ram", "Biratnagar", "22"],
    ["Shyam", "Itahari", "24"],
    ["Hari", "Dharan", "23"],
]

additional_data = ['Himal', "Dharan" , "23"]
with open(file="newnew.csv" , mode='w' , newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
    
    writer.writerow(additional_data)
```

```python
data = [
    {"name": "Himal", "age": 30},
    {"name": "Sajina", "age": 10},
    {"name": "Heemal", "age": 33},
]
with open(file="output.csv", mode="w") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "age"])
    writer.writeheader()
    writer.writerows(data)
```

```python
data = [
    {"Roll": 1, "Name": "Ram", "English": 80, "Math": 81, "Science": 82},
    {"Roll": 2, "Name": "Shyam", "English": 79, "Math": 80, "Science": 83},
]
dataframe = pd.DataFrame(data=data)  # total vanne chaiyo. and then percentage vanne banaune.

dataframe["Total"] = dataframe["English"] + dataframe["Math"] + dataframe["Science"]
print(dataframe)
```

```python
dataframe['Percentage'] =(dataframe['Total']/3).round(decimals=2)
print(dataframe)
```

```python
dataframe.to_csv("newFile.csv", index=False)
```
